```python
import numpy as np
import matplotlib.pyplot as plt
series = np.loadtxt('Taxi_Test.txt')
series = series.reshape(-1,1)
print(series.shape)
plt.figure()
n=500
length = 288
plt.plot(range(length), series[n:n+length,-1], label='1')
plt.plot(range(length), series[n+length:n+length*2,-1], label='2')
plt.plot(range(length), series[n+length*2:n+length*3,-1], label='3')
plt.plot(range(length), series[n+length*3:n+length*4,-1], label='4')
plt.plot(range(length), series[n+length*4:n+length*5,-1], label='5')
plt.legend()
plt.show()
```
